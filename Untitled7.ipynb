```python
# Print i as long as i is less than 6:
i=1
while i<=5:
  print(i)
  i+=1
```

```python
# print numbers from one to n
n=int(input("Enter a number: "))
i=1
while i<=n:
  print(i)
  i+=1
```

```python
#factorial of a number
n=int(input("Enter a number: "))
fact=1
if n<0:
  print("Factorial does not exist for negative numbers")
else:
  while n>0:
    fact*=n
    n-=1
  print("factorial=",fact)
```

```python
#Fibonacci Series using while loop
n=int(input("Enter a number: "))
a=0
b=1
i=1
while i<=n:
  print(a,end=" ")
  c=a+b
  a=b
  b=c
  i+=1
```

```python
#To limit the value of sequence to set number
limit=int(input("Enter a limit: "))
a=0
b=1
while a<=limit:
  print(a,end=" ")
  a,b=b,a+b
```

```python
#Reverse a number
num=int(input("enter your number: "))
rev=0
while num >0:
  digit=num % 10
  rev=rev * 10 + digit
  num //=10
print("reverse number is: ",rev)
```

```python
# Check Palidrome number /string
num=int(input("Enter a number: "))
temp=num
rev=0
while num>0:
  rev = rev * 10 + num % 10
  num//=10
if temp==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
s="madam"
i,j=0,len(s)-1
is_palindrome=True
while i<j:
  if s[i]!=s[j]:
    is_palindrome=False
    break
  i+=1
  j-=1
if is_palindrome:
  print("Yes")
else:
  print("No")
```

```python
#check palindrome number
st=input("enter the string")
rev=st[::-1]
if st==rev:
  print("palindrome")
else:
  print("not palindrome")
```

```python
#count digits in a number
num=int(input("Enter a number: "))
count=0
while num>0:
  num//=10
  count+=1
print("No of digits",count)
```

```python
#exit the loop
i=1
while i<6:
  print(i)
  if i==3:
    break
  i+=1
```

```python
#search an elements in a list
nums=[10,20,30,40,50]
key=int (input("Enter element to search"))
i=0
while i<len(nums):
  if nums[i]==key:
    print("Element found at index",i)
    print("Element found at position",i+1)
    break
  i+=1
else:
    print("element not found")
```
